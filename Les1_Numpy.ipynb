###Задание_1

```
import numpy as np
```

```
a = np.array([[1, 6], 
             [2, 8], 
             [3, 11], 
             [3, 10], 
             [1, 7]])
```

```
a.shape
```

```
mean_a = a.mean(axis = 0)
mean_a
```

###Задание_2

```
a_centered = a - mean_a
a_centered
```

```
a_centered.shape
```

###Задание_3

```
a1 = a_centered[:, 0]
a2 = a_centered[:, 1]
a_centered_sp = np.dot(a1, a2)
a_centered_sp
```

```
N = a.shape[0]
N
```

```
b = a_centered_sp/(N-1)
b
```
